#### Tarefa 2. Descubra qual a idade média por ocupação.

*Dica: use a função groupby do Pandas na coluna occupation*

````python
import pandas as pd

users = pd.read_csv(r"./datasets/users_dataset.csv", sep="|")
````

#### Tarefa 3. Para cada ocupação descubra a idade mínima e máxima.

````python
users[['occupation', 'age']].groupby("occupation", as_index=False).agg({'age': ['min', 'max']})
````

#### Tarefa 4. Para cada combinação de ocupação e sexo, calcule a idade média.

Exemplo:

```
occupation     gender
administrator  F         40.638889
               M         37.162791
artist         F         30.307692
               M         32.333333
```

````python
users[['occupation', 'age', 'gender']].groupby(["occupation", "gender"]).mean()
````

#### Tarefa 5. Para cada ocupação calcule a porcentagem de homens e mulheres.

Exemplo:

```
occupation     gender
administrator  F          45.569620
               M          54.430380
artist         F          46.428571
               M          53.571429
doctor         M         100.000000
educator       F          27.368421
               M          72.631579
```

````python
temp = users.groupby(["occupation"])["gender"].value_counts()

pct = temp.groupby(level=0).apply(lambda x: 100 * x / float(x.sum()))
pct
````
